```python
import requests
import fitz  # PyMuPDF
from lxml import etree

# Função para baixar o PDF de uma URL
def download_pdf(url, save_path):
    response = requests.get(url)
    if response.status_code == 200:
        with open(save_path, 'wb') as f:
            f.write(response.content)
    else:
        print(f"Failed to download PDF. Status code: {response.status_code}")

# Função para extrair texto de um PDF
def extract_text_from_pdf(pdf_path):
    document = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text()
    return text

# Função para criar um arquivo XML a partir do texto extraído
def create_xml_from_text(text, xml_path):
    # Criando o elemento raiz
    root = etree.Element("document")

    # Dividindo o texto em linhas
    lines = text.split('\n')

    # Adicionando cada linha como um elemento XML
    for line in lines:
        if line.strip():  # Ignorando linhas vazias
            element = etree.Element("line")
            element.text = line.strip()
            root.append(element)

    # Escrevendo o XML para o arquivo
    tree = etree.ElementTree(root)
    tree.write(xml_path, pretty_print=True, xml_declaration=True, encoding="UTF-8")

# URL do PDF
pdf_url = 'https://ainfo.cnptia.embrapa.br/digital/bitstream/item/105942/1/Doc256-OL.pdf'

# Caminho para salvar o PDF baixado
pdf_path = 'downloaded_file.pdf'

# Caminho para o arquivo XML que será criado
xml_path = 'output.xml'

# Baixando o PDF
download_pdf(pdf_url, pdf_path)

# Extraindo texto do PDF
text = extract_text_from_pdf(pdf_path)

# Criando o XML a partir do texto extraído
create_xml_from_text(text, xml_path)

print(f'Arquivo XML criado com sucesso em {xml_path}')



```

```python
from tika import parser
from lxml import etree
# Função para extrair texto de um arquivo PDF usando Tika
def extract_text_from_pdf(pdf_path):
    raw = parser.from_file(pdf_path)
    return raw['content']


# Caminho para o arquivo de texto de entrada
input_path = 'cnptia2.txt'

# Caminho para o arquivo de texto de saída
output_path = 'filtered_cnptia2.txt'

# Caminho para o arquivo PDF
pdf_path = 'downloaded_file.pdf'

# Extraindo texto do PDF
text = extract_text_from_pdf(pdf_path)

# Imprimindo o texto extraído
print(text)

with open(input_path, 'w') as f:
  f.write(text)

def read_text_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def filter_paragraphs_by_keywords(text, keywords):
    # Dividir o texto em parágrafos. Supondo que os parágrafos são separados por duas quebras de linha.
    paragraphs = text.split('\n\n')
    filtered_paragraphs = []

    for paragraph in paragraphs:
        if any(keyword.lower() in paragraph.lower() for keyword in keywords):
            paragraph = paragraph.replace('\n', ' ')  # Removendo quebras de linha
            filtered_paragraphs.append(paragraph)

    return filtered_paragraphs

def save_filtered_paragraphs(filtered_paragraphs, output_path):
    with open(output_path, 'w', encoding='utf-8') as file:
        for paragraph in filtered_paragraphs:
            file.write(paragraph + '\n\n')  # Adicionando duas quebras de linha para separar os parágrafos.

# Lista de palavras-chave
keywords = ['soja']

# Lendo o texto do arquivo
text = read_text_file(input_path)

# Filtrando os parágrafos pelas palavras-chave
filtered_paragraphs = filter_paragraphs_by_keywords(text, keywords)

# Salvando os parágrafos filtrados em um novo arquivo
save_filtered_paragraphs(filtered_paragraphs, output_path)

print(f'Parágrafos filtrados salvos em {output_path}')

```

```python
# prompt: faça pra mim um código csv com o arquivo agrofyInsumos.txt buscando a div que procure a tag span title=

import csv
from bs4 import BeautifulSoup

# Abrindo o arquivo HTML
with open('agrofyInsumos.txt', 'r') as file:
    html_content = file.read()

# Criando o objeto BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Encontrando todas as tags span com o atributo title
spans_with_title = soup.find_all('span', title=True)

# Criando uma lista para armazenar os dados
data = []

# Iterando sobre as tags span e extraindo o título e o texto
for span in spans_with_title:
    title = span['title']
    data.append([title])

# Criando o arquivo CSV
with open('agrofy_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Título'])  # Escrevendo o cabeçalho
    writer.writerows(data)  # Escrevendo os dados

print('Arquivo CSV "agrofy_data.csv" criado com sucesso.')
```

```python
# prompt: faça um código montando um html para os arquivos agrofy_data.csv, climatempo.csv, embrapanoticias.csv, filtered_cnptia2.csv

def create_html_from_csv(csv_files):
  html_content = """
  <!DOCTYPE html>
  <html>
  <head>
  <title>Dados de Arquivos CSV</title>
  <style>
  table {
    border-collapse: collapse;
    width: 100%;
  }
  th, td {
    border: 1px solid black;
    padding: 8px;
    text-align: left;
  }
  </style>
  </head>
  <body>
  """

  for csv_file in csv_files:
    try:
      with open(csv_file, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        header = next(reader)  # Lê a primeira linha como cabeçalho
        html_content += f"<h2>{csv_file}</h2>"
        html_content += "<table>"
        html_content += "<tr>" + "".join(f"<th>{col}</th>" for col in header) + "</tr>"
        for row in reader:
          html_content += "<tr>" + "".join(f"<td>{col}</td>" for col in row) + "</tr>"
        html_content += "</table><br>"
    except FileNotFoundError:
      html_content += f"<p>Arquivo {csv_file} não encontrado.</p><br>"

  html_content += """
  </body>
  </html>
  """
  return html_content

csv_files = ['agrofy_data.csv', 'climatempo.csv', 'embrapanoticias.csv', 'filtered_cnptia2.csv']
html_output = create_html_from_csv(csv_files)

with open('dados_csv.html', 'w', encoding='utf-8') as f:
  f.write(html_output)

print('Arquivo HTML "dados_csv.html" criado com sucesso.')
```
